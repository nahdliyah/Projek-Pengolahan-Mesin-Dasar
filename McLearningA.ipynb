```python
selected_columns = ['Age', 'Gender', 'Air Pollution', 'Alcohol use', 'Dust Allergy', 'OccuPational Hazards', 'Genetic Risk', 'chronic Lung Disease', 'Balanced Diet', 'Obesity', 'Smoking', 'Passive Smoker', 'Chest Pain', 'Coughing of Blood', 'Fatigue', 'Weight Loss', 'Shortness of Breath', 'Wheezing', 'Swallowing Difficulty', 'Clubbing of Finger Nails', 'Frequent Cold', 'Dry Cough', 'Snoring', 'Level']
databaru = data[selected_columns]
databaru.head()
```

**PREPOSESSING**

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
nilai_min = databaru.min()
nilai_max = databaru.max()
print("Nilai Minimum dari Setiap Kolom:")
print(nilai_min)

print("\nNilai Maksimum dari Setiap Kolom:")
print(nilai_max)
```

```python
scaler = MinMaxScaler()
scaled_data = scaler.fit_transform(databaru.drop('Level', axis=1))
```

```python
# Langkah 1: Impor Pustaka yang Diperlukan
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt


# Tambahkan kembali kolom target ke df_scaled
df_scaled['Level'] = databaru['Level'].values

# Langkah 3: Hitung Korelasi antara Semua Fitur dan Target
correlation_matrix = df_scaled.corr()

# Ambil korelasi dengan kolom target
correlation_with_target = correlation_matrix['Level'].sort_values(ascending=False)

# Tampilkan korelasi dengan kolom target
print("Korelasi dengan kolom target:")
print(correlation_with_target)

# Langkah 4: Visualisasikan Korelasi
plt.figure(figsize=(12, 10))
sns.heatmap(correlation_matrix, annot=True, fmt='.2f', cmap='coolwarm')
plt.title('Heatmap Korelasi Fitur')
plt.show()

# Langkah 5: Pilih Fitur Berdasarkan Korelasi
# Tentukan threshold (batas) korelasi untuk memilih fitur
threshold = 0.5

# Pilih fitur yang memiliki korelasi absolut lebih besar dari threshold dengan target
selected_features = correlation_with_target[abs(correlation_with_target) > threshold].index.tolist()

# Hapus target column dari daftar fitur terpilih
if 'Level' in selected_features:
    selected_features.remove('Level')

print(f'Fitur terpilih berdasarkan korelasi (threshold={threshold}): {selected_features}')

# Dataset baru dengan fitur terpilih
df_selected = df_scaled[selected_features]

# Periksa lima baris pertama dataset baru dengan fitur terpilih
df_selected
```

```python
df_selected.shape
```

```python
# Gabungkan DataFrame hasil korelasi dengan DataFrame target berdasarkan indeks
data_select = pd.concat([df_selected, databaru['Level']], axis=1)

# Tampilkan DataFrame hasil gabungan
data_select
```

```python
import pandas as pd
import matplotlib.pyplot as plt

target_counts = data_select['Level'].value_counts()
# Buat pie chart
plt.figure(figsize=(8, 6))
plt.pie(target_counts, labels=target_counts.index, autopct='%1.1f%%', startangle=140)
plt.title('Pie Chart Berdasarkan Target')
plt.axis('equal')  # Memastikan lingkaran menjadi lingkaran, bukan elips
plt.show()
```

```python
from sklearn.model_selection import train_test_split

# Bagi dataset menjadi data latih dan data uji
X_train, X_test, y_train, y_test = train_test_split(data_select.drop('Level', axis=1), data_select['Level'], test_size=0.2, random_state=42)
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.naive_bayes import GaussianNB
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn import model_selection
from sklearn.svm import LinearSVC

# Membuat array kosong untuk menampung beberapa algoritma klasifikasi
models = []

# Membuat array asosiatif berisi nama algoritma dan algoritma klasifikasinya
models.append(('Decision Tree', DecisionTreeClassifier()))
models.append(('Random Forest', RandomForestClassifier()))
models.append(('SVM', SVC()))
models.append(('Logistic Regression', LogisticRegression()))
models.append(('KNN', KNeighborsClassifier(n_neighbors=7, metric='euclidean')))
models.append(('Linear SVM', LinearSVC(dual=True)))


seed = 7
scoring = 'accuracy'
results = []

# Membagi data menjadi training dan validation set menggunakan k-fold cross-validation, dengan k = 7
for name, model in models:
    kfold = model_selection.KFold(n_splits=7, random_state=seed, shuffle=True)

    # Menguji akurasi dari masing-masing model menggunakan validasi data
    cv_results = model_selection.cross_val_score(model, X_train, y_train.ravel(), cv=kfold, scoring=scoring)
    results.append(cv_results)
    msg = "%s : %f (%f)" % (name, cv_results.mean(), cv_results.std())
    print(msg)
```

###Pemodelan

```python
from sklearn.svm import SVC
classifier = SVC()
classifier.fit(X_train,y_train)
Y_predd = classifier.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

import scikitplot as skplt
cm = confusion_matrix(y_test,Y_predd)
print(cm)
print("Akurasi:", accuracy_score(y_test, Y_predd))
print("Laporan Klasifikasi:\n", classification_report(y_test, Y_predd))
print("Confussion Matriks:\n", confusion_matrix(y_test, Y_predd))
skplt.metrics.plot_confusion_matrix(y_test, Y_predd, normalize=False, title = 'Confussion Matriks',figsize=(5,3))
```
